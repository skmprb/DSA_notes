```python
from typing import List

class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        nums_dict = dict()

        for i in range(len(nums)):
            if nums[i] not in nums_dict:
                nums_dict[nums[i]] = self.subProduct(nums, i)
        
        output = []
        for n in nums:
            output.append(nums_dict[n])
        
        return output

    
    def subProduct(self, nums, i):
        result = 1
        for j in range(len(nums)):
            if i == j: continue
            result *= nums[j]
        
        return result
```
