```python
def maxminum_product_subarrat(nums):
    n = len(nums)
    max_prod = nums[0]

    for i in range (n):
        current_prod = 1
        for j in range(i,n):
            current_prod *= nums[j]

            max_prod = max(current_prod, max_prod)
    
    return max_prod
print(maxminum_product_subarrat([-2,3,-4]))



# Approach 1: Track curr_max and curr_min at each position
# Why curr_min? Negative numbers can flip min to max
def max_product(nums):
    n = len(nums)
    curr_max = nums[0]
    curr_min = nums[0]
    max_prod = nums[0]

    for i in range(1, n):
        # Consider: start fresh, extend max, extend min (for negatives)
        temp = max(nums[i], curr_max*nums[i], curr_min*nums[i])
        curr_min = min(nums[i], curr_max*nums[i], curr_min*nums[i])
        curr_max = temp
        
        max_prod = max(max_prod, curr_max)
    
    return max_prod

print(max_product([2, 3, -2, 4]))  # Output: 6 ([2,3])


# Return subarray with maximum product
def max_product_with_subarray(nums):
    n = len(nums)
    curr_max = nums[0]
    curr_min = nums[0]
    max_prod = nums[0]
    
    start = 0
    end = 0
    temp_start = 0
    
    for i in range(1, n):
        # Check if starting fresh gives better result
        if nums[i] > curr_max * nums[i] and nums[i] > curr_min * nums[i]:
            curr_max = nums[i]
            curr_min = nums[i]
            temp_start = i
        else:
            temp_max = max(curr_max * nums[i], curr_min * nums[i])
            temp_min = min(curr_max * nums[i], curr_min * nums[i])
            curr_max = temp_max
            curr_min = temp_min
        
        if curr_max > max_prod:
            max_prod = curr_max
            start = temp_start
            end = i
    
    return nums[start:end+1], max_prod

subarray, product = max_product_with_subarray([2, 3, -2, 4])
print(f"Subarray: {subarray}, Product: {product}")


# Approach 2: Swap curr_max and curr_min when encountering negative
# Simpler logic - negative flips max/min relationship
def max_product_swap(nums):
    global_max = float('-inf')
    curr_max = 1
    curr_min = 1

    for num in nums:
        if num < 0:  # Swap when negative
            curr_max, curr_min = curr_min, curr_max
        
        curr_max = max(num, num * curr_max)
        curr_min = min(num, num * curr_min)
        global_max = max(curr_max, global_max)
    
    return global_max


# Approach 3: Two-pass (left-to-right, right-to-left)
# Handles zeros and negatives by scanning both directions
def max_product_two_pass(nums):
    ans = float("-inf")
    curr_prod = 1

    # Left to right
    for num in nums:
        if curr_prod == 0:
            curr_prod = 1
        curr_prod *= num
        ans = max(curr_prod, ans)
    
    # Right to left
    curr_prod = 1
    for i in range(len(nums) - 1, -1, -1):
        if curr_prod == 0:
            curr_prod = 1
        curr_prod *= nums[i]
        ans = max(curr_prod, ans)
    
    return ans


# Approach 4: Handle zeros explicitly by resetting
def max_product_zero_reset(nums):
    res = max(nums)
    curmax, curmin = 1, 1
    
    for i in nums:
        if i == 0:  # Reset on zero
            curmin, curmax = 1, 1
            continue
        
        temp = curmax * i
        curmax = max(curmax*i, curmin*i, i)
        curmin = min(temp, curmin*i, i)
        res = max(res, curmax)
    
    return res         


```
