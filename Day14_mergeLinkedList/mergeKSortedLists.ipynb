You are given an array of k linked-lists lists, each linked-list is sorted in ascending order.

Merge all the linked-lists into one sorted linked-list and return it.

 

Example 1:

Input: lists = [[1,4,5],[1,3,4],[2,6]]
Output: [1,1,2,3,4,4,5,6]
Explanation: The linked-lists are:
[
  1->4->5,
  1->3->4,
  2->6
]
merging them into one sorted linked list:
1->1->2->3->4->4->5->6
Example 2:

Input: lists = []
Output: []
Example 3:

Input: lists = [[]]
Output: []
 

Constraints:

k == lists.length
0 <= k <= 104
0 <= lists[i].length <= 500
-104 <= lists[i][j] <= 104
lists[i] is sorted in ascending order.
The sum of lists[i].length will not exceed 104.

```python
# Definition for singly-linked list.
from typing import Optional, List

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:

    def mergeTwoLists(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        res = ListNode(0)
        cur = res
        while l1 and l2:
            if l1.val < l2.val:
                cur.next = l1
                l1 = l1.next
            else:
                cur.next = l2
                l2 = l2.next
            cur = cur.next
        
        cur.next = l1 or l2
        return res.next

    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        if not lists:
            return None
        if len(lists) == 1:
            return lists[0]
        
        mid = len(lists) // 2
        # Bug: Missing 'self.' prefix for recursive calls to mergeKLists
        left = self.mergeKLists(lists[:mid])
        right = self.mergeKLists(lists[mid:])
        # Bug: Missing 'self.' prefix for call to mergeTwoLists
        return self.mergeTwoLists(left,right)

        
```

```python
# This function creates linked lists from arrays of values
# It takes variable number of array arguments (*arrays)
# Returns a single linked list if one array is provided, or a list of linked lists if multiple arrays are provided
def create_list(*arrays):
    # If no arrays are provided, return None
    if not arrays:
        return None
    
    # Initialize list to store the head nodes of created linked lists
    heads = []
    for arr in arrays:
        # If the current array is empty, append None to heads list
        if not arr:
            heads.append(None)
            continue
        # Create the first node of the linked list with the first array element
        head = ListNode(arr[0])
        current = head
        # Create remaining nodes by iterating through the rest of the array
        for val in arr[1:]:
            current.next = ListNode(val)
            current = current.next
        # Add the head of this linked list to the heads list
        heads.append(head)
    
    # Return list of heads if multiple arrays were provided, otherwise return single head
    return heads if len(heads) > 1 else heads[0]

# This function converts linked list(s) back to array format for easy visualization
def list_to_array(head):
    # Handle case when head is a list of linked lists
    if isinstance(head, list):
        return [list_to_array(h) for h in head]
    
    # Convert single linked list to array
    result = []
    while head:
        result.append(head.val)
        head = head.next
    return result

# Create an instance of Solution class (assumed to have mergeKLists method)
sol = Solution()

# Test case 1: Create linked lists from arrays [1,2,3,4,5], [3,4,7], [1,3,4]
# This creates 3 separate linked lists and returns them as a list
head1 = create_list([1,2,3,4,5],[3,4,7],[1,3,4])
# Merge the k linked lists using Solution's mergeKLists method and convert result to array for printing
print(list_to_array(sol.mergeKLists(head1)))

# Test case 2: Create linked lists from two empty arrays
# This creates 2 None values (empty linked lists)
head2 = create_list([],[])
# Merge the empty linked lists and convert result to array for printing
print(list_to_array(sol.mergeKLists(head2)))
```

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

class PriorityWrapper:
    def __init__(self, val, node):
        self.val = val
        self.node = node

    def __lt__(self, other):
        return self.val < other.val

from queue import PriorityQueue
class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        pq = PriorityQueue()

        for node in lists:
            if node:
                pq.put(PriorityWrapper(node.val, node))
        
        head = ListNode(-1, None)
        curr = head

        while not pq.empty():
            pw = pq.get()
            val, node = pw.val, pw.node
            curr.next = node
            curr = node

            if node.next:
                pq.put(PriorityWrapper(node.next.val, node.next))
        
        return head.next
```

```python
class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        if not lists:
            return None
        
        def merge2lists(l1, l2):
            dummy = ListNode(0)
            cur = dummy
            
            while l1 and l2:
                if l1.val <= l2.val:
                    cur.next = l1
                    l1 = l1.next
                else:
                    cur.next = l2
                    l2 = l2.next
                cur = cur.next
            
            cur.next = l1 if l1 else l2
            return dummy.next
        
        head = None
        
        for l in lists:
            head = merge2lists(head, l)
        
        return head
```
