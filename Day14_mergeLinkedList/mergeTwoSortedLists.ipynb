You are given the heads of two sorted linked lists list1 and list2.

Merge the two lists into one sorted list. The list should be made by splicing together the nodes of the first two lists.

Return the head of the merged linked list.

 

Example 1:


Input: list1 = [1,2,4], list2 = [1,3,4]
Output: [1,1,2,3,4,4]
Example 2:

Input: list1 = [], list2 = []
Output: []
Example 3:

Input: list1 = [], list2 = [0]
Output: [0]
 

Constraints:

The number of nodes in both lists is in the range [0, 50].
-100 <= Node.val <= 100
Both list1 and list2 are sorted in non-decreasing order.

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
from typing import Optional


class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        # Step 1: Create a dummy node to simplify the merging process
        # This dummy node will be the starting point of our merged list
        dummy = ListNode(0)
        # Step 2: Initialize current pointer to track the last node in our merged list
        current = dummy
        # Step 3: Loop while both lists have remaining nodes
        while list1 and list2:
            # Step 4: Compare the values of the current nodes in both lists
            if list1.val < list2.val:
                # Step 5a: If list1's current node has smaller value, add it to merged list
                # Bug fix: should assign list1 node, not list1.next
                current.next = list1
                # Step 6a: Move list1 pointer to the next node
                list1 = list1.next
            else:
                # Step 5b: If list2's current node has smaller or equal value, add it to merged list
                # Bug fix: should assign list2 node, not list2.next
                current.next = list2
                # Step 6b: Move list2 pointer to the next node
                list2 = list2.next
            # Step 7: Move current pointer to the newly added node
            current = current.next
        # Step 8: After the loop, one list might still have remaining nodes
        # Append the remaining nodes (if any) to the merged list
        current.next = list1 or  list2
        # Step 9: Return the merged list starting from the node after dummy
        # (dummy.next is the actual head of the merged list)
        return dummy.next

```
