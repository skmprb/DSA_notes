You are given a string s and an integer k. You can choose any character of the string and change it to any other uppercase English character. You can perform this operation at most k times.

Return the length of the longest substring containing the same letter you can get after performing the above operations.

 

Example 1:

Input: s = "ABAB", k = 2
Output: 4
Explanation: Replace the two 'A's with two 'B's or vice versa.
Example 2:

Input: s = "AABABBA", k = 1
Output: 4
Explanation: Replace the one 'A' in the middle with 'B' and form "AABBBBA".
The substring "BBBB" has the longest repeating letters, which is 4.
There may exists other ways to achieve this answer too.
 

Constraints:

1 <= s.length <= 105
s consists of only uppercase English letters.
0 <= k <= s.length

```python
def characterReplacement(s,k):
    left = 0
    max_length = 0
    count = {}
    max_freq = 0

    for right in range(len(s)):

        # print(f"before {count.get(s[right])}")
        count[s[right]] = count.get(s[right],0)+1
        # print(f"after {count[s[right]]}")
        # print(count)

        max_freq = max(max_freq, count[s[right]])

        while(right-left +1) - max_freq > k:
            count[s[left]] -= 1
            left += 1
        
        max_length = max(max_length, right-left+1)
    
    return max_length

print(characterReplacement("AABABBA",1))
```

```python
class Solution:
    def characterReplacement(self, s: str, k: int) -> int:
        freq = {}
        left = 0
        max_freq = 0
        replacements_used = 0

        for char in s:
            freq[char] = freq.get(char, 0) + 1

            if freq[char] > max_freq:
                max_freq = freq[char]
            elif replacements_used < k:
                replacements_used += 1
            else:
                freq[s[left]] -= 1
                left += 1

        return max_freq + replacements_used
```

## Optimized Solutions

### Key Optimization: Don't update max_freq when shrinking
- We only care about finding LARGER windows
- If max_freq decreases, window won't grow anyway
- This saves recalculating max(count.values())

```python
# Optimized Approach 1: Using array instead of dict
def characterReplacement_array(s, k):
    count = [0] * 26
    left = 0
    max_length = 0
    max_freq = 0
    
    for right in range(len(s)):
        idx = ord(s[right]) - ord('A')
        count[idx] += 1
        max_freq = max(max_freq, count[idx])
        
        if (right - left + 1) - max_freq > k:
            count[ord(s[left]) - ord('A')] -= 1
            left += 1
        
        max_length = max(max_length, right - left + 1)
    
    return max_length

print(characterReplacement_array("ABAB", 2))
print(characterReplacement_array("AABABBA", 1))
```

```python
# Optimized Approach 2: With early termination
def characterReplacement_optimized(s, k):
    n = len(s)
    if k >= n - 1:
        return n
    
    count = {}
    left = 0
    max_length = 0
    max_freq = 0
    
    for right in range(n):
        count[s[right]] = count.get(s[right], 0) + 1
        max_freq = max(max_freq, count[s[right]])
        
        if (right - left + 1) - max_freq > k:
            count[s[left]] -= 1
            left += 1
        
        max_length = max(max_length, right - left + 1)
    
    return max_length

print(characterReplacement_optimized("ABAB", 2))
print(characterReplacement_optimized("AABABBA", 1))
```
