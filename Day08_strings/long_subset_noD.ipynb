Given a string s, find the length of the longest substring without duplicate characters.

 

Example 1:

Input: s = "abcabcbb"
Output: 3
Explanation: The answer is "abc", with the length of 3. Note that "bca" and "cab" are also correct answers.
Example 2:

Input: s = "bbbbb"
Output: 1
Explanation: The answer is "b", with the length of 1.
Example 3:

Input: s = "pwwkew"
Output: 3
Explanation: The answer is "wke", with the length of 3.
Notice that the answer must be a substring, "pwke" is a subsequence and not a substring.
 

Constraints:

0 <= s.length <= 5 * 104
s consists of English letters, digits, symbols and spaces.

```python
def lengthOfLongestSubstring(s):

    char_set = set()
    left = 0
    max_length = 0

    for right in range(len(s)):

        # print(right, s[right], left)
        while s[right] in char_set:
            char_set.remove(s[left])
            # print(char_set)
            left +=1
            # print(left)

        # print(s[right])
        char_set.add(s[right])
        max_length = max(max_length, right-left +1)
    return max_length

print(lengthOfLongestSubstring("abcabcbb"))

```

```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        if len(s) == 0:
            return 0
        
        res = 1
        left = 0
        right = 1 # right != left

        while right < len(s):
            if s[right] not in s[left: right]:
                temp = right - left + 1
                if temp > res:
                    res = temp
                right += 1
            else:
                for i in range(left,right):
                    if s[i] == s[right]:
                        left = i + 1
                        continue
        
        return res
```

## Optimized Solutions

### Approach Comparison:
1. **Set-based** (above): O(2n) time - each char visited twice
2. **HashMap-based** (above): O(n) time - direct jump
3. **ASCII Array** (below): O(n) time, faster constant
4. **Bit Manipulation** (below): O(n) time, for lowercase only

```python
# Optimized Approach 1: Using ASCII array (fastest for ASCII)
def lengthOfLongestSubstring_ascii(s):
    """
    Time: O(n), Space: O(1) - fixed 128 size
    Best for ASCII characters
    """
    char_index = [-1] * 128  # ASCII character set
    left = 0
    max_length = 0
    
    for right in range(len(s)):
        # Get ASCII value of character
        ascii_val = ord(s[right])
        
        # If character seen and within current window
        if char_index[ascii_val] >= left:
            left = char_index[ascii_val] + 1
        
        # Update index and max length
        char_index[ascii_val] = right
        max_length = max(max_length, right - left + 1)
    
    return max_length

# Test
print(lengthOfLongestSubstring_ascii("abcabcbb"))  # 3
print(lengthOfLongestSubstring_ascii("bbbbb"))     # 1
print(lengthOfLongestSubstring_ascii("pwwkew"))    # 3
```

```python
# Optimized Approach 2: Bit manipulation (for lowercase letters only)
def lengthOfLongestSubstring_bits(s):
    """
    Time: O(n), Space: O(1)
    Only works for lowercase English letters (a-z)
    Uses bitmask for O(1) lookup
    """
    left = 0
    max_length = 0
    char_bits = 0  # Bitmask to track characters
    
    for right in range(len(s)):
        char_bit = 1 << (ord(s[right]) - ord('a'))
        
        # While character already in window
        while char_bits & char_bit:
            # Remove leftmost character
            char_bits ^= 1 << (ord(s[left]) - ord('a'))
            left += 1
        
        # Add current character
        char_bits |= char_bit
        max_length = max(max_length, right - left + 1)
    
    return max_length

# Test (only lowercase)
print(lengthOfLongestSubstring_bits("abcabcbb"))  # 3
print(lengthOfLongestSubstring_bits("bbbbb"))     # 1
```

```python
# Optimized Approach 3: Early termination
def lengthOfLongestSubstring_optimized(s):
    """
    Time: O(n), Space: O(min(n, m))
    With early termination optimization
    """
    n = len(s)
    
    # Early termination
    if n == 0:
        return 0
    if n == 1:
        return 1
    
    # If remaining chars can't beat current max, stop
    char_index = {}
    left = 0
    max_length = 0
    
    for right in range(n):
        # Early termination: if remaining length can't beat max
        if n - left <= max_length:
            break
        
        if s[right] in char_index and char_index[s[right]] >= left:
            left = char_index[s[right]] + 1
        
        char_index[s[right]] = right
        max_length = max(max_length, right - left + 1)
    
    return max_length

# Test
print(lengthOfLongestSubstring_optimized("abcabcbb"))  # 3
print(lengthOfLongestSubstring_optimized("bbbbb"))     # 1
print(lengthOfLongestSubstring_optimized("pwwkew"))    # 3
```

## Performance Comparison

| Approach | Time | Space | Best For |
|----------|------|-------|----------|
| Set-based | O(2n) | O(min(n,m)) | Simple implementation |
| HashMap | O(n) | O(min(n,m)) | General case |
| ASCII Array | O(n) | O(1) | ASCII characters (fastest) |
| Bit Manipulation | O(n) | O(1) | Lowercase letters only |
| Early Termination | O(n) | O(min(n,m)) | Large strings with early patterns |
