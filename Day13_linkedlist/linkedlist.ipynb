```python
# Node class represents a single element in the linked list
class Node:
    def __init__(self, data):
        # Store the actual data/value in this node
        self.data = data
        # Pointer to the next node in the list (initially None)
        self.next = None

# LinkedList class manages the entire linked list structure
class LinkedList:
    def __init__(self):
        # self.head is a reference/pointer to the FIRST node in the linked list
        # It marks the beginning/start of the entire linked list
        # When the list is empty, self.head is None
        # When the list has elements, self.head points to the first node
        self.head = None

    # Method to display all elements in the linked list - O(n)
    def display(self):
        elements = []
        current = self.head
        while current:
            elements.append(f"{current.data}(id:{id(current)})")
            current = current.next
        print(" -> ".join(elements) + " -> None")

    # Insert at beginning - O(1) time complexity
    def insert_at_beginning(self, data):
        print(f"\n--- INSERT AT BEGINNING: {data} ---")
        print(f"Before: ", end="")
        self.display()
        
        # Step 1: Create a new node with the given data
        new_node = Node(data)
        print(f"Step 1: Created new node with data={data}")
        
        # Step 2: Point the new node's next to the current head
        new_node.next = self.head
        print(f"Step 2: New node's next now points to current head ({self.head.data if self.head else 'None'})")
        
        # Step 3: Update head to point to the new node (making it the first node)
        self.head = new_node
        print(f"Step 3: Head updated to new node")
        
        print(f"After: ", end="")
        self.display()

    # Insert at end - O(n) time complexity
    def insert_at_end(self, data):
        print(f"\n--- INSERT AT END: {data} ---")
        print(f"Before: ", end="")
        self.display()
        
        # Step 1: Create a new node with the given data
        new_node = Node(data)
        print(f"Step 1: Created new node with data={data}")
        
        # Step 2: Check if list is empty
        if not self.head:
            self.head = new_node
            print(f"Step 2: List was empty, new node is now head")
            print(f"After: ", end="")
            self.display()
            return
        
        # Step 3: Traverse to find the last node
        current = self.head
        print(f"Step 2: Starting traversal from head (data={current.data})")
        position = 0
        while current.next:
            print(f"  Position {position}: current={current.data}, moving to next")
            current = current.next
            position += 1
        print(f"  Position {position}: current={current.data} (last node found)")
        
        # Step 4: Link the last node to the new node
        current.next = new_node
        print(f"Step 3: Linked last node to new node")
        
        print(f"After: ", end="")
        self.display()

    # Insert at specific position - O(n)
    def insert_at_position(self, data, position):
        print(f"\n--- INSERT AT POSITION {position}: {data} ---")
        print(f"Before: ", end="")
        self.display()
        
        # Step 1: Check if inserting at beginning
        if position == 0:
            print(f"Step 1: Position is 0, calling insert_at_beginning")
            self.insert_at_beginning(data)
            return
        
        # Step 2: Create new node
        new_node = Node(data)
        print(f"Step 1: Created new node with data={data}")
        
        # Step 3: Traverse to position-1
        current = self.head
        print(f"Step 2: Traversing to position {position-1}")
        for i in range(position - 1):
            if not current:
                print(f"  Error: Position {position} is out of bounds")
                return
            print(f"  At index {i}: current={current.data}")
            current = current.next
        
        # Step 4: Insert new node
        print(f"Step 3: At position {position-1}, current={current.data}")
        new_node.next = current.next
        print(f"Step 4: New node's next points to {current.next.data if current.next else 'None'}")
        current.next = new_node
        print(f"Step 5: Current node's next now points to new node")
        
        print(f"After: ", end="")
        self.display()

    # Delete from beginning - O(1)
    def delete_from_beginning(self):
        print(f"\n--- DELETE FROM BEGINNING ---")
        print(f"Before: ", end="")
        self.display()
        
        # Step 1: Check if list is empty
        if not self.head:
            print(f"Step 1: List is empty, nothing to delete")
            return
        
        # Step 2: Store the node to be deleted
        deleted_data = self.head.data
        print(f"Step 1: Deleting node with data={deleted_data}")
        
        # Step 3: Move head to next node
        self.head = self.head.next
        print(f"Step 2: Head now points to {self.head.data if self.head else 'None'}")
        
        print(f"After: ", end="")
        self.display()

    # Delete from end - O(n)
    def delete_from_end(self):
        print(f"\n--- DELETE FROM END ---")
        print(f"Before: ", end="")
        self.display()
        
        # Step 1: Check if list is empty
        if not self.head:
            print(f"Step 1: List is empty, nothing to delete")
            return
        
        # Step 2: Check if only one node exists
        if not self.head.next:
            deleted_data = self.head.data
            print(f"Step 1: Only one node with data={deleted_data}, removing it")
            self.head = None
            print(f"Step 2: Head is now None")
            print(f"After: ", end="")
            self.display()
            return
        
        # Step 3: Traverse to second-to-last node
        current = self.head
        print(f"Step 1: Traversing to second-to-last node")
        position = 0
        while current.next.next:
            print(f"  Position {position}: current={current.data}, next={current.next.data}")
            current = current.next
            position += 1
        
        # Step 4: Remove last node
        deleted_data = current.next.data
        print(f"Step 2: At second-to-last node (data={current.data}), deleting last node (data={deleted_data})")
        current.next = None
        print(f"Step 3: Set current.next to None")
        
        print(f"After: ", end="")
        self.display()

    # Delete by value - O(n)
    def delete_by_value(self, value):
        print(f"\n--- DELETE BY VALUE: {value} ---")
        print(f"Before: ", end="")
        self.display()
        
        # Step 1: Check if list is empty
        if not self.head:
            print(f"Step 1: List is empty, nothing to delete")
            return
        
        # Step 2: Check if head node has the value
        if self.head.data == value:
            print(f"Step 1: Head node has value={value}, deleting it")
            self.head = self.head.next
            print(f"Step 2: Head now points to {self.head.data if self.head else 'None'}")
            print(f"After: ", end="")
            self.display()
            return
        
        # Step 3: Traverse to find the node with the value
        current = self.head
        print(f"Step 1: Searching for value={value}")
        position = 0
        while current.next:
            print(f"  Position {position}: current={current.data}, next={current.next.data}")
            if current.next.data == value:
                print(f"Step 2: Found value at position {position+1}")
                current.next = current.next.next
                print(f"Step 3: Removed node, current.next now points to {current.next.data if current.next else 'None'}")
                print(f"After: ", end="")
                self.display()
                return
            current = current.next
            position += 1
        
        print(f"Value {value} not found in list")

    # Search - O(n)
    def search(self, value):
        print(f"\n--- SEARCH FOR VALUE: {value} ---")
        current = self.head
        position = 0
        # Traverse through the list
        while current:
            print(f"Position {position}: current={current.data}")
            if current.data == value:
                print(f"Found {value} at position {position}")
                return position
            current = current.next
            position += 1
        print(f"Value {value} not found")
        return -1

    # Get length - O(n)
    def length(self):
        print(f"\n--- GET LENGTH ---")
        count = 0
        current = self.head
        # Count each node
        while current:
            print(f"Position {count}: current={current.data}")
            count += 1
            current = current.next
        print(f"Total length: {count}")
        return count

    # Reverse - O(n)
    def reverse(self):
        print(f"\n--- REVERSE LIST ---")
        print(f"Before: ", end="")
        self.display()
        
        # Initialize pointers
        prev = None
        current = self.head
        step = 1
        
        # Reverse the links
        while current:
            print(f"Step {step}: current={current.data}, prev={prev.data if prev else 'None'}")
            # Save next node
            next_node = current.next
            print(f"  Saved next={next_node.data if next_node else 'None'}")
            # Reverse the link
            current.next = prev
            print(f"  Reversed: current.next now points to {prev.data if prev else 'None'}")
            # Move pointers forward
            prev = current
            current = next_node
            step += 1
        
        # Update head to the last node (which is now first)
        self.head = prev
        print(f"Final: Head updated to {self.head.data if self.head else 'None'}")
        
        print(f"After: ", end="")
        self.display()

    # Get middle element - O(n)
    def get_middle(self):
        print(f"\n--- GET MIDDLE ELEMENT ---")
        # Use slow and fast pointer technique
        slow = fast = self.head
        step = 0
        
        # Fast moves 2 steps, slow moves 1 step
        while fast and fast.next:
            print(f"Step {step}: slow={slow.data}, fast={fast.data}")
            slow = slow.next
            fast = fast.next.next
            step += 1
        
        result = slow.data if slow else None
        print(f"Middle element: {result}")
        return result

    # Detect cycle - O(n)
    def has_cycle(self):
        print(f"\n--- DETECT CYCLE ---")
        # Use slow and fast pointer technique
        slow = fast = self.head
        step = 0
        
        # If fast catches slow, there's a cycle
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
            print(f"Step {step}: slow={slow.data}, fast={fast.data if fast else 'None'}")
            if slow == fast:
                print(f"Cycle detected!")
                return True
            step += 1
        
        print(f"No cycle detected")
        return False
```

```python
# Test the LinkedList with detailed output
ll = LinkedList()

# Test insert operations
ll.insert_at_beginning(5)
ll.insert_at_beginning(3)
ll.insert_at_end(7)
ll.insert_at_position(4, 1)

# Test search and length
ll.search(4)
ll.length()

# Test delete operations
ll.delete_by_value(4)
ll.delete_from_beginning()
ll.delete_from_end()

# Test reverse and middle
ll.insert_at_end(10)
ll.insert_at_end(20)
ll.get_middle()
ll.reverse()

# Test cycle detection
ll.has_cycle()
```
