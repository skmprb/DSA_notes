Design a stack that supports push, pop, top, and retrieving the minimum element in constant time.

Implement the MinStack class:

MinStack() initializes the stack object.
void push(int val) pushes the element val onto the stack.
void pop() removes the element on the top of the stack.
int top() gets the top element of the stack.
int getMin() retrieves the minimum element in the stack.
You must implement a solution with O(1) time complexity for each function.

 

Example 1:

Input
["MinStack","push","push","push","getMin","pop","top","getMin"]
[[],[-2],[0],[-3],[],[],[],[]]

Output
[null,null,null,null,-3,null,0,-2]

Explanation
MinStack minStack = new MinStack();
minStack.push(-2);
minStack.push(0);
minStack.push(-3);
minStack.getMin(); // return -3
minStack.pop();
minStack.top();    // return 0
minStack.getMin(); // return -2
 

Constraints:

-231 <= val <= 231 - 1
Methods pop, top and getMin operations will always be called on non-empty stacks.
At most 3 * 104 calls will be made to push, pop, top, and getMin.

```python
class MinStack:

    def __init__(self):
        # Initialize the main stack to store all values
        self.stack = []
        # Initialize auxiliary stack to keep track of minimum values at each level
        self.minstack = []
        

    def push(self, val: int) -> None:
        # Add the new value to the main stack
        self.stack.append(val)
        
        # If minstack is empty or new value is smaller/equal to current minimum
        if not self.minstack or val <= self.minstack[-1]:
            # Push the new value as it becomes the new minimum
            self.minstack.append(val)
        else:
            # Push the current minimum again to maintain stack alignment
            # This ensures minstack always has same length as main stack
            self.minstack.append(self.minstack[-1])

    def pop(self) -> None:
        # Check if stack is empty to avoid errors
        if not self.stack:
            return
        # Remove the top element from minstack to maintain synchronization
        self.minstack.pop()
        # Remove and return the top element from main stack
        return self.stack.pop()

    
    def top(self) -> int:
        # Check if stack is empty before accessing top element
        if not self.stack:
            return
        
        # Return the top element without removing it
        return self.stack[-1]
        

    def getMin(self) -> int:
        # Check if minstack is empty (though this should match main stack)
        if not self.minstack:
            return -1
        # Return the current minimum which is always at top of minstack
        return self.minstack[-1]

    def print(self) -> int:
        # Debug function to visualize both stacks
        print(self.stack)
        print(self.minstack)

# Create an instance of MinStack for testing
st = MinStack()
# Push -2: stack=[-2], minstack=[-2]
st.push(-2)
# Push 0: stack=[-2,0], minstack=[-2,-2] (0 > -2, so repeat -2)
st.push(0)
# Push -3: stack=[-2,0,-3], minstack=[-2,-2,-3] (-3 < -2, so new min)
st.push(-3)
# Print current state of both stacks
st.print()
# Get minimum value (-3)
print(st.getMin())
# Pop top element (-3) and return it
print(st.pop())
# Print state after pop: stack=[-2,0], minstack=[-2,-2]
st.print()
# Get top element (0) without removing
print(st.top())
# Print unchanged state
st.print()
# Get current minimum (-2)
print(st.getMin())
# Print final state
st.print()
        


# Your MinStack object will be instantiated and called as such:
# obj = MinStack()
# obj.push(val)
# obj.pop()
# param_3 = obj.top()
# param_4 = obj.getMin()
```

```python
class MinStack:

    def __init__(self):
        self.stack = []
        self.minstack = []
        

    def push(self, val: int) -> None:
        self.stack.append(val)
        
        if not self.minstack or val <= self.minstack[-1]:
            self.minstack.append(val)
        else:
            self.minstack.append(self.minstack[-1])

    def pop(self) -> None:
        if not self.stack:
            return
        self.minstack.pop()
        return self.stack.pop()

    
    def top(self) -> int:
        if not self.stack:
            return
        
        return self.stack[-1]
        

    def getMin(self) -> int:
        if not self.minstack:
            return -1
        return self.minstack[-1]

    def print(self) -> int:
        print(self.stack)
        print(self.minstack)

st = MinStack()
st.push(-2)
st.push(0)
st.push(-3)
st.print()
print(st.getMin())
print(st.pop())
st.print()
print(st.top())
st.print()
print(st.getMin())
st.print()
        


# Your MinStack object will be instantiated and called as such:
# obj = MinStack()
# obj.push(val)
# obj.pop()
# param_3 = obj.top()
# param_4 = obj.getMin()
```
