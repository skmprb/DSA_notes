Given an array of integers temperatures represents the daily temperatures, return an array answer such that answer[i] is the number of days you have to wait after the ith day to get a warmer temperature. If there is no future day for which this is possible, keep answer[i] == 0 instead.

 

Example 1:

Input: temperatures = [73,74,75,71,69,72,76,73]
Output: [1,1,4,2,1,1,0,0]
Example 2:

Input: temperatures = [30,40,50,60]
Output: [1,1,1,0]
Example 3:

Input: temperatures = [30,60,90]
Output: [1,1,0]

```python
# Monotonic Stack (decreasing) algorithm
def DailyTemperature(nums) : 
    answers = [0] * len(nums)
    stack = []

    for i , t in enumerate(nums):
        while stack and stack[-1][0] < t:
            stack_temp, stack_index = stack.pop()
            answers[stack_index] = i - stack_index
        stack.append((t,i))
    
    return answers

print(DailyTemperature([73,74,75,71,69,72,76,73]))
print(DailyTemperature([30,40,50,60]))
print(DailyTemperature([30, 60, 90]))

```

```python
from typing import List
class Solution:
    def dailyTemperatures(self, arr: List[int]) -> List[int]:
        st=[]
        res=[0]*len(arr)
        for i in range(len(arr)-1 , -1, -1):
            while st and arr[st[-1]]<=arr[i]:
                st.pop()
            if st:
                res[i]=st[-1]-i
            st.append(i)
        return res                
                 
```

```python
# solution with no stack

def dailytempe(nums):
    n = len(nums)
    answers = [0] * n
    hottest = 0

    for i in range(n-1, -1,-1):
        if nums[i] >= hottest:
            hottest = nums[i]
            continue
        days = 1
        while nums[i+days] < nums[i]:
            days += answers[i+days]
        answers[i] = days
    
    return answers

print(dailytempe([73,74,75,71,69,72,76,73]))
```
