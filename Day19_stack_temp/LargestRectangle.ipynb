Given an array of integers heights representing the histogram's bar height where the width of each bar is 1, return the area of the largest rectangle in the histogram.

 

Example 1:


Input: heights = [2,1,5,6,2,3]
Output: 10
Explanation: The above is a histogram where width of each bar is 1.
The largest rectangle is shown in the red area, which has an area = 10 units.
Example 2:


Input: heights = [2,4]
Output: 4
 

Constraints:

1 <= heights.length <= 105
0 <= heights[i] <= 104

```python
def largeRectangle(heights):
    n = len(heights)
    max_area = 0
    stack = []

    for i, height in enumerate(heights):
        start = i

        while stack and stack[-1][0] > height:
            h, j = stack.pop()
            w = i - j
            max_area = max(max_area, h*w)
            start = j

        stack.append((height, start))

    while stack:
        h,j = stack.pop()
        w= n - j
        max_area = max(max_area, h*w)
    return max_area

print (largeRectangle([2,1,5,6,2,3]))
```

```python
from typing import List
class Solution:
    def largestRectangleArea(self, h: List[int]) -> int:
        res = 0
        # Initialize the maximum area found so far
        l = []
        # Stack to store indices of histogram bars
        h.append(0)
        # Add a sentinel value (0) to ensure all bars are processed
        n = len(h)
        # Get the length of the modified histogram
        for i in range(n):
        
        # Iterate through each bar in the histogram
            while l and h[l[-1]] > h[i]:
            # While stack is not empty and current bar is shorter than the bar at stack top
                height = h[l.pop()]
                # Pop the top index and use it as the height of rectangle
                width = i if not l else i - l[-1] - 1
                # Calculate width: if stack empty, width is current index
                # Otherwise, width is distance between current index and new stack top
                area = height * width
                # Calculate area of rectangle with this height
                res = max(res,area)
                # Update maximum area if current area is larger
                res = max(res, area)
            # Push current index onto stack
        return res
        
    def largestRectangleAreaOptimized(self, heights: List[int]) -> int:
        # Alternative implementation using cleaner variable names
        max_area = 0
        stack = []
        
        # Process each bar including a sentinel 0 at the end
        for i in range(len(heights) + 1):
            # Current height is 0 if we're at the end, otherwise get from array
            current_height = 0 if i == len(heights) else heights[i]
            
            # Process all bars in stack that are taller than current bar
            while stack and heights[stack[-1]] > current_height:
                # The height of rectangle is the height of the popped bar
                rect_height = heights[stack.pop()]
                # Width calculation: distance to current position minus distance to previous bar
                rect_width = i if not stack else i - stack[-1] - 1
                # Calculate and update maximum area
                max_area = max(max_area, rect_height * rect_width)
            
            # Add current index to stack
            stack.append(i)
        
        return max_area

```

```python
class Solution:
    def largestRectangleArea(self, heights: List[int]) -> int:
        # Initialize stack with -1 as a sentinel value to help calculate widths
        stack = [-1]
        # Track the maximum rectangle area found so far
        max_area = 0
        
        # Iterate through each bar in the histogram
        for i in range(len(heights)):
            # While stack is not empty and current bar is shorter than the bar at stack top
            # Pop bars from stack and calculate area with popped bar as the smallest
            while stack[-1] != -1 and heights[stack[-1]] >= heights[i]:
                # Height of the rectangle is the height of the popped bar
                curr_height = heights[stack.pop()]
                # Width is from the bar after the new stack top to current position
                curr_width = i - stack[-1] - 1
                # Update maximum area if current area is larger
                max_area = max(max_area,curr_height*curr_width)
            # Push current bar index onto stack
            stack.append(i)
        
        # Process remaining bars in stack (bars that extend to the end)
        while stack[-1] != -1:
            # Height of rectangle is the height of the popped bar
            curr_height = heights[stack.pop()]
            # Width extends from after the new stack top to the end of array
            curr_width = len(heights) - stack[-1] - 1
            # Update maximum area if current area is larger
            max_area = max(max_area,curr_height*curr_width)
        
        # Return the largest rectangle area found
        return max_area
```
