## Technique 1: Two Pointers (Converging)

### When to Use:
- **Verify** if something IS a palindrome
- **Check** validity of a string/array
- You have a **fixed range** to check

### Core Concept:
```
Start from both ends → Move towards center → Compare characters

   "r a c e c a r"
    ↑           ↑
   left       right
    
   "r a c e c a r"
      ↑       ↑
     left   right
     
   "r a c e c a r"
        ↑   ↑
      left right
```

### Template:

````python
# TEMPLATE: Two Pointers for Palindrome Check
def is_palindrome(s, left, right):
    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1
    return True
````

### Example 1: Valid Palindrome (Basic)

````python
# Problem: Check if string is palindrome (ignore non-alphanumeric, case-insensitive)
# Input: "A man, a plan, a canal: Panama"
# Output: True

def isPalindrome(s):
    left, right = 0, len(s) - 1
    
    while left < right:
        # Skip non-alphanumeric from left
        while left < right and not s[left].isalnum():
            left += 1
        
        # Skip non-alphanumeric from right
        while left < right and not s[right].isalnum():
            right -= 1
        
        # Compare (case-insensitive)
        if s[left].lower() != s[right].lower():
            return False
        
        left += 1
        right -= 1
    
    return True

# Test
print(isPalindrome("A man, a plan, a canal: Panama"))  # True
print(isPalindrome("race a car"))  # False
````

### Example 2: Valid Palindrome II (Delete at most 1 character)

````python
# Problem: Can string become palindrome by deleting at most 1 character?
# Input: "abca"
# Output: True (delete 'c' or 'b')

def validPalindrome(s):
    def check(left, right):
        while left < right:
            if s[left] != s[right]:
                return False
            left += 1
            right -= 1
        return True
    
    left, right = 0, len(s) - 1
    
    while left < right:
        if s[left] != s[right]:
            # Try skipping left OR right character
            return check(left + 1, right) or check(left, right - 1)
        left += 1
        right -= 1
    
    return True

# Test
print(validPalindrome("abca"))  # True
print(validPalindrome("abc"))   # False
````

## Technique 2: Expand Around Center

### When to Use:
- **Find** palindromes within a string
- **Count** how many palindromes exist
- **Get the longest** palindrome
- You need to **explore all possibilities**

### Core Concept:
```
Pick a center → Expand outward → Check if characters match

Odd length (center = 1 char):     Even length (center = 2 chars):
      "a b a"                            "a b b a"
        ↑                                  ↑ ↑
      center                            center
      
    ← a b a →                          ← a b b a →
      expand                              expand
```

### Why Check Both Odd and Even?
- **Odd**: "aba" has center 'b'
- **Even**: "abba" has center between 'b' and 'b'

### Template:

````python
# TEMPLATE: Expand Around Center
def expand_around_center(s, left, right):
    while left >= 0 and right < len(s) and s[left] == s[right]:
        # Do something with palindrome s[left:right+1]
        left -= 1
        right += 1
    return left + 1, right - 1  # Return valid indices

# For each position in string:
for i in range(len(s)):
    # Check odd length palindrome (center = i)
    expand_around_center(s, i, i)
    
    # Check even length palindrome (center = i and i+1)
    expand_around_center(s, i, i + 1)
````

### Example 1: Longest Palindromic Substring

````python
# Problem: Find the longest palindromic substring
# Input: "babad"
# Output: "bab" or "aba"

def longestPalindrome(s):
    def expand(left, right):
        # Expand while characters match
        while left >= 0 and right < len(s) and s[left] == s[right]:
            left -= 1
            right += 1
        # Return the palindrome substring
        return s[left + 1:right]
    
    result = ""
    
    for i in range(len(s)):
        # Check odd length
        odd = expand(i, i)
        # Check even length
        even = expand(i, i + 1)
        # Keep the longest
        result = max(result, odd, even, key=len)
    
    return result

# Test
print(longestPalindrome("babad"))  # "bab" or "aba"
print(longestPalindrome("cbbd"))   # "bb"
````

### Example 2: Palindromic Substrings (Count)

````python
# Problem: Count all palindromic substrings
# Input: "aaa"
# Output: 6 ("a", "a", "a", "aa", "aa", "aaa")

def countSubstrings(s):
    def expand(left, right):
        count = 0
        # Count each palindrome found while expanding
        while left >= 0 and right < len(s) and s[left] == s[right]:
            count += 1
            left -= 1
            right += 1
        return count
    
    total = 0
    
    for i in range(len(s)):
        # Count odd length palindromes
        total += expand(i, i)
        # Count even length palindromes
        total += expand(i, i + 1)
    
    return total

# Test
print(countSubstrings("aaa"))  # 6
print(countSubstrings("abc"))  # 3
````

## Decision Tree: Which Technique to Use?

```
Question about palindrome?
    |
    ├─ Need to CHECK/VERIFY if something IS palindrome?
    |     → Use TWO POINTERS (converging)
    |     → Start from both ends, move to center
    |
    └─ Need to FIND/COUNT palindromes?
          → Use EXPAND AROUND CENTER
          → Try each position as center, expand outward
```

## Practice Problems

### Practice 1: Shortest Palindrome (Hard)
Add minimum characters to front to make palindrome

````python
# Problem: Add minimum chars to front to make palindrome
# Input: "aacecaaa"
# Output: "aaacecaaa"

def shortestPalindrome(s):
    # Find longest palindrome starting from index 0
    def is_palindrome(left, right):
        while left < right:
            if s[left] != s[right]:
                return False
            left += 1
            right -= 1
        return True
    
    # Find longest palindrome prefix
    for i in range(len(s) - 1, -1, -1):
        if is_palindrome(0, i):
            # Add reverse of remaining suffix to front
            return s[i + 1:][::-1] + s
    
    return s

# Test
print(shortestPalindrome("aacecaaa"))  # "aaacecaaa"
print(shortestPalindrome("abcd"))      # "dcbabcd"
````

### Practice 2: Longest Palindrome by Concatenating (Medium)

````python
# Problem: Find longest palindrome length from array of strings
# Input: ["lc","cl","gg"]
# Output: 6 ("lc" + "gg" + "cl" = "lcggcl")

def longestPalindrome(words):
    from collections import Counter
    count = Counter(words)
    length = 0
    has_center = False
    
    for word in count:
        reverse = word[::-1]
        
        if word == reverse:  # Palindrome word like "gg"
            if count[word] % 2 == 0:
                length += count[word] * 2
            else:
                length += (count[word] - 1) * 2
                has_center = True
        elif reverse in count and word < reverse:  # Pair like "lc" and "cl"
            length += min(count[word], count[reverse]) * 4
    
    return length + (2 if has_center else 0)

# Test
print(longestPalindrome(["lc","cl","gg"]))  # 6
print(longestPalindrome(["ab","ty","yt","lc","cl","ab"]))  # 8
````

## Key Takeaways

### Two Pointers (Converging):
- **Purpose**: Verify/Check
- **Direction**: Outside → Inside
- **Time**: O(n)
- **Keywords**: "is palindrome", "valid", "check"

### Expand Around Center:
- **Purpose**: Find/Count
- **Direction**: Inside → Outside
- **Time**: O(n²)
- **Keywords**: "longest", "count", "find all"
- **Remember**: Always check BOTH odd and even lengths!

### Common Mistakes:
1. Forgetting to check even-length palindromes
2. Off-by-one errors in indices
3. Not handling edge cases (empty string, single character)
4. Forgetting to skip non-alphanumeric in validation problems
