Given a string s, return the longest palindromic substring in s.

 

Example 1:

Input: s = "babad"
Output: "bab"
Explanation: "aba" is also a valid answer.
Example 2:

Input: s = "cbbd"
Output: "bb"
 

Constraints:

1 <= s.length <= 1000
s consist of only digits and English letters.

```python
# Approach 1: Expand Around Center (Optimized)
def longestpalindrome(s):
    n = len(s)
    start, maxlent = 0, 1  # Track starting index and max length of palindrome
    
    for i in range(n):  # Iterate through each character as potential center
        for j in range(2):  # j=0 for odd length, j=1 for even length palindromes
            low, high = i, i+j  # Initialize pointers
            
            # Expand outward while characters match
            while low >= 0 and high < n and s[low] == s[high]:
                currentlength = high - low + 1
                
                # Update if we found a longer palindrome
                if currentlength > maxlent:
                    start = low
                    maxlent = currentlength
                
                low -= 1  # Move left pointer outward
                high += 1  # Move right pointer outward
    
    return s[start:start+maxlent]

print(longestpalindrome("babad"))
print(longestpalindrome("cbbd"))
print(longestpalindrome(""))
```

```python
# Approach 2: Brute Force - Check all substrings
def check(str, low, hight):
    # Helper function to verify if substring is palindrome
    while low < hight:
        if str[low] != str[hight]:
            return False  # Characters don't match
        low += 1
        hight -= 1
    return True  # All characters matched

def getLongestpal(s):
    n = len(s)
    maxlen = 1  # Minimum palindrome length is 1
    start = 0  # Starting index of longest palindrome found
    
    # Try all possible substrings
    for i in range(n):
        for j in range(i, n):
            # Check if substring is palindrome and longer than current max
            if check(s, i, j) and (j - i + 1) > maxlen:
                maxlen = j - i + 1
                start = i
    
    return s[start: start+maxlen]

print(getLongestpal("babad"))
```

```python
# Approach 3: Expand Around Center (Detailed version)

def longestPalindrome(s):
    res = ""  # Store the longest palindrome found
    res_len = 0  # Track length of longest palindrome

    for i in range(len(s)):  # Each position as potential center

        # Check odd length palindromes (single character center)
        l = r = i  # Both pointers start at same position
        while l >= 0 and r < len(s) and s[l] == s[r]:
            # Update result if current palindrome is longer
            if (r - l + 1) > res_len:
                res = s[l:r+1]
                res_len = r - l + 1
            l -= 1  # Expand left
            r += 1  # Expand right

        # Check even length palindromes (two character center)
        l, r = i, i+1  # Pointers start at adjacent positions
        while l >= 0 and r < len(s) and s[l] == s[r]:
            # Update result if current palindrome is longer
            if (r - l + 1) > res_len:
                res = s[l:r+1]
                res_len = r - l + 1
            l -= 1  # Expand left
            r += 1  # Expand right

    return res

print(longestPalindrome("babad"))
```

```python
# Approach 4: Cleaner version using helper function

def longestpalindrome(s):
    def expand(l, r):
        # Expand outward while characters match
        while l >= 0 and r < len(s) and s[l] == s[r]:
            l -= 1
            r += 1
        # Return the palindrome substring (l+1 because l went one step too far)
        return s[l+1:r]
    
    res = ""  # Store longest palindrome
    for i in range(len(s)):
        # Compare current result with odd and even length palindromes
        # expand(i, i) checks odd length, expand(i, i+1) checks even length
        res = max(res, expand(i, i), expand(i, i+1), key=len)
    return res

print(longestpalindrome("babad"))
```

```python
# Approach 5: Brute Force with String Reversal
class Solution:
    def longestPalindrome(self, s: str) -> str:
        ans = s[0]  # Initialize with first character
        maxLen = 1  # Minimum length is 1
        
        # Try all possible starting positions
        for i in range(0, len(s)-1):
            currString = s[i]  # Start with single character
            currLen = 1
            
            # Extend substring by adding characters
            for j in range(i+1, len(s)):
                currString += s[j]
                currLen += 1
                
                # Check if palindrome by comparing with reverse
                if currString == currString[::-1] and currLen > maxLen:
                    maxLen = currLen
                    ans = currString

        return ans
```

```python
# Approach 6: Optimized Expand Around Center with Index Tracking
class Solution:
    def longestPalindrome(self, s: str) -> str:
        n = len(s)
        if n < 1:
            return s

        def expand(l: int, r: int) -> tuple[int, int]:
            # Expand while characters match and return final indices
            while l >= 0 and r < n and s[l] == s[r]:
                l -= 1
                r += 1
            # Return valid indices (adjust for overshoot)
            return l + 1, r - 1

        best_l, best_r = 0, 0  # Track best palindrome indices

        for i in range(n):  # Try each position as center
            # Check odd length palindrome (single center)
            l1, r1 = expand(i, i)
            if r1 - l1 > best_r - best_l:
                best_l, best_r = l1, r1

            # Check even length palindrome (double center)
            l2, r2 = expand(i, i + 1)
            if r2 - l2 > best_r - best_l:
                best_l, best_r = l2, r2

        return s[best_l:best_r + 1]
```
