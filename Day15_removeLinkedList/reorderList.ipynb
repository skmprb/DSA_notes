You are given the head of a singly linked-list. The list can be represented as:

L0 → L1 → … → Ln - 1 → Ln
Reorder the list to be on the following form:

L0 → Ln → L1 → Ln - 1 → L2 → Ln - 2 → …
You may not modify the values in the list's nodes. Only nodes themselves may be changed.

 

Example 1:


Input: head = [1,2,3,4]
Output: [1,4,2,3]
Example 2:


Input: head = [1,2,3,4,5]
Output: [1,5,2,4,3]
 

Constraints:

The number of nodes in the list is in the range [1, 5 * 104].
1 <= Node.val <= 1000

```python
from typing import Optional, List
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def reorderList(self, head: Optional[ListNode]) -> None:
        """
        Do not return anything, modify head in-place instead.
        """

        if not head or not head.next :
            return None
        
        # middle number
        fast = slow = head
        while fast.next and fast.next.next:
            slow = slow.next
            fast = fast.next.next
        
        #reverse second
        second = slow.next
        slow.next = None
        prev = None
        while second:
            next_node = second.next
            second.next = prev
            prev = second
            second = next_node
        
        # merge firat and second

        first, second = head, prev
        while second:
            temp1, temp2 = first.next, second.next
            first.next = second
            second.next = temp1
            first, second = temp1, temp2
        
```

```python
def create_list(arr):
    if not arr:
        return None
    head = ListNode(arr[0])
    current = head
    for val in arr[1:]:
        current.next = ListNode(val)
        current = current.next
    return head

def list_to_array(head):
    result = []
    while head:
        result.append(head.val)
        head = head.next
    return result

sol = Solution()

head1 = create_list([1,2,3,4,5])
sol.reorderList(head1)
print(list_to_array(head1))

head2 = create_list([1])
sol.reorderList(head2)
print(list_to_array(head2))

head3 = create_list([1,2])
sol.reorderList(head3)
print(list_to_array(head3))

head4 = create_list([])
sol.reorderList(head4)
print(list_to_array(head4))

```

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

# This code implements a linked list reordering algorithm using a "move last to front" pattern
# It repeatedly takes the last node and moves it to the position after the current left pointer
class Solution:
    def reorderList(self, head: Optional[ListNode]) -> None:
        """
        Do not return anything, modify head in-place instead.
        """
        # Initialize left pointer starting from head
        l = head
        # Continue while there are at least 2 more nodes after current left pointer
        while l.next and l.next.next:
            # Initialize right pointer to traverse from current left position
            r = l
            # Traverser right pointer to the end of list
            while r.next:
                # Check if we've reached the second-to-last node
                if not r.next.next:
                    # Store reference to second-to-last node
                    temp = r
                    # Move r to the last node
                    r = r.next
                    # Disconnect the last node from the list
                    temp.next = None
                else:
                    # Continue traversing to the right
                    r = r.next


            # Insert the last node (r) between l and l.next
            # Connect last node to the node that was after l
            r.next = l.next
            # Connect l to the last node
            l.next = r
            # Move left pointer two positions forward (skip the inserted node)
            l = l.next.next
        return

        
```

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

# This code implements a three-step pattern for reordering a linked list:
# 1. Find middle and split list, 2. Collect second half values, 3. Interleave values from end
class Solution:
    def reorderList(self, head: Optional[ListNode]) -> None:
        """
        Do not return anything, modify head in-place instead.
        """
        # Array to store values from second half of the list
        new = []
        # Two pointers to traverse the list
        first = head
        second = head
        # Counter to find the length of the list
        count = 0
        
        # First pass: count total nodes in the list
        while first:
            count += 1
            first = first.next
        
        # Calculate midpoint - if even length use count/2, if odd use count//2 + 1
        if count % 2 == 0:
            mid = count / 2
        else:
            mid = count // 2 + 1
        
        # Reset counter and initialize previous pointer
        count = 0
        prev = None
        
        # Second pass: split list at midpoint and collect second half values
        while second:
            if count >= mid:
                # Cut the connection to split the list
                prev.next = None
                # Store values from second half in array
                new.append(second.val)
            else:
                # Keep track of previous node to cut connection later
                prev = second
            second = second.next
            count += 1

        # Third pass: interleave nodes by inserting values from end of second half
        while new:
            # Save the next node in original first half
            next = head.next
            # Create new node with value from end of second half (pop removes last element)
            head.next = ListNode(new.pop())
            # Connect the new node to the rest of original first half
            head.next.next = next
            # Move to next position in first half
            head = next
```
