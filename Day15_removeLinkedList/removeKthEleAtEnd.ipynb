Given the head of a linked list, remove the nth node from the end of the list and return its head.

 

Example 1:


Input: head = [1,2,3,4,5], n = 2
Output: [1,2,3,5]
Example 2:

Input: head = [1], n = 1
Output: []
Example 3:

Input: head = [1,2], n = 1
Output: [1]
 

Constraints:

The number of nodes in the list is sz.
1 <= sz <= 30
0 <= Node.val <= 100
1 <= n <= sz
 

Follow up: Could you do this in one pass?

```python
from typing import Optional , List
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        dummy = ListNode(0)
        dummy.next = head
        slow = fast = dummy

        for _ in range(n+1):
            if not fast:
                return None
            fast = fast.next
        
        while fast:
            fast = fast.next
            slow = slow.next
        
        slow.next = slow.next.next
        return dummy.next
```

```python
def create_list(arr):
    if not arr:
        return None
    head = ListNode(arr[0])
    current = head
    for val in arr[1:]:
        current.next = ListNode(val)
        current = current.next
    return head

def list_to_array(head):
    result = []
    while head:
        result.append(head.val)
        head = head.next
    return result

sol = Solution()

head1 = create_list([1,2,3,4,5])
k = 2
print(list_to_array(sol.removeNthFromEnd(head1,k)))

head2 = create_list([1])
k = 2
print(list_to_array(sol.removeNthFromEnd(head2,k)))

head3 = create_list([1,2])
k = 1
print(list_to_array(sol.removeNthFromEnd(head3,k)))

```

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

# This solution uses the Two Pointer technique (fast and slow pointers)
# Pattern: Two-pass approach with pointer manipulation to find the nth node from end
class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        # Initialize two pointers starting at head
        slow = head
        fast = head
        # Move fast pointer n-1 steps ahead to create gap of n nodes
        for _ in range(n - 1):
            fast = fast.next
        # Keep track of the node before slow (for deletion)
        pre = slow
        # Move both pointers until fast reaches the end
        while fast and fast.next:
            fast = fast.next
            pre = slow  # Update previous node before moving slow
            slow = slow.next
        # Remove the nth node by updating previous node's next pointer
        pre.next = slow.next
        # Handle edge case: if removing the head node
        if slow == head:
            return slow.next
        # Return the modified list
        return head
        
```
