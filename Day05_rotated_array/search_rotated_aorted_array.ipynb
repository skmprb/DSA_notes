There is an integer array nums sorted in ascending order (with distinct values).

Prior to being passed to your function, nums is possibly left rotated at an unknown index k (1 <= k < nums.length) such that the resulting array is [nums[k], nums[k+1], ..., nums[n-1], nums[0], nums[1], ..., nums[k-1]] (0-indexed). For example, [0,1,2,4,5,6,7] might be left rotated by 3 indices and become [4,5,6,7,0,1,2].

Given the array nums after the possible rotation and an integer target, return the index of target if it is in nums, or -1 if it is not in nums.

You must write an algorithm with O(log n) runtime complexity.

 

Example 1:

Input: nums = [4,5,6,7,0,1,2], target = 0
Output: 4
Example 2:

Input: nums = [4,5,6,7,0,1,2], target = 3
Output: -1
Example 3:

Input: nums = [1], target = 0
Output: -1
 

Constraints:

1 <= nums.length <= 5000
-104 <= nums[i] <= 104
All values of nums are unique.
nums is an ascending array that is possibly rotated.
-104 <= target <= 104

````python
def search_key(nums,target):
    left = 0
    right = len(nums)-1

    while left <= right:

        mid = left + (right - left)//2

        if nums[mid] == target:
            return mid

        if nums[mid] >= nums[left]:

            if nums[left] <= target and nums[mid] >= target:
                right = mid-1
            else:
                left = mid+1
        else:
            if nums[right] >= target and nums[mid] <= target:
                left = mid+1
            else:
                right = mid - 1
    return -1

print(search_key([4,5,6,7,0,1,2], 0))
print(search_key([1], 1))
print(search_key([4,5,6,7,0,1,2], 3))
print(search_key([1,3], 3))
print(search_key([5,1,3], 3))


````

````python
def search_key(nums, target):
    left = 0
    right = len(nums)-1

    while left <= right:

        mid = left + (right-left)//2

        if nums[mid] == target:
            return mid

        if nums[mid] >= nums[left]:

            if target >= nums[left] and target < nums[mid]:
                right = mid - 1
            else:
                left = mid + 1
        else:
            if target > nums[mid] and target <= nums[right]:
                left = mid + 1
            else:
                right = mid - 1
    return -1

print(search_key([4,5,6,7,0,1,2], 0))
print(search_key([1], 1))
print(search_key([4,5,6,7,0,1,2], 3))
print(search_key([1,3], 3))
print(search_key([5,1,3], 3))
````

## Comparison of Two Approaches

Both solutions are correct but use different comparison logic:

### Approach 1: Check if target is OUTSIDE sorted range
```python
if nums[left] <= target and nums[mid] >= target:  # target IN left sorted range
    right = mid - 1
else:  # target OUTSIDE left range, search right
    left = mid + 1
```

### Approach 2: Check if target is INSIDE sorted range
```python
if target >= nums[left] and target < nums[mid]:  # target IN left sorted range
    right = mid - 1
else:  # target OUTSIDE left range, search right
    left = mid + 1
```

### Key Difference

**Approach 1**: Uses `<=` and `>=` (inclusive on both ends)
- `nums[left] <= target and nums[mid] >= target`
- Checks if target is within [nums[left], nums[mid]]

**Approach 2**: Uses `>=` and `<` (inclusive left, exclusive right)
- `target >= nums[left] and target < nums[mid]`
- Checks if target is within [nums[left], nums[mid])

### Why Both Work

Since we already checked `nums[mid] == target` at the start:
- Approach 1: `nums[mid] >= target` will never be true when `nums[mid] == target`
- Approach 2: `target < nums[mid]` explicitly excludes mid

Both correctly identify if target is in the sorted portion!

### Example: [4,5,6,7,0,1,2], target = 0

**Step 1**: left=0, right=6, mid=3, nums[mid]=7
- Left half [4,5,6,7] is sorted
- Approach 1: `4 <= 0` is FALSE → search right
- Approach 2: `0 >= 4` is FALSE → search right
- Both go right ✓

**Step 2**: left=4, right=6, mid=5, nums[mid]=1
- Right half [1,2] is sorted (nums[mid] < nums[left])
- Approach 1: `2 >= 0 and 1 <= 0` is FALSE → search left
- Approach 2: `0 > 1 and 0 <= 2` is FALSE → search left
- Both go left ✓

**Conclusion**: Both approaches are equivalent and correct. Choose based on preference!
- Approach 1: More symmetric (both use <=, >=)
- Approach 2: More explicit (excludes mid with <)
