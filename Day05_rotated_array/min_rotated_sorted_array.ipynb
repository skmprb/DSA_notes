Suppose an array of length n sorted in ascending order is rotated between 1 and n times. For example, the array nums = [0,1,2,4,5,6,7] might become:

[4,5,6,7,0,1,2] if it was rotated 4 times.
[0,1,2,4,5,6,7] if it was rotated 7 times.
Notice that rotating an array [a[0], a[1], a[2], ..., a[n-1]] 1 time results in the array [a[n-1], a[0], a[1], a[2], ..., a[n-2]].

Given the sorted rotated array nums of unique elements, return the minimum element of this array.

You must write an algorithm that runs in O(log n) time.

 

Example 1:

Input: nums = [3,4,5,1,2]
Output: 1
Explanation: The original array was [1,2,3,4,5] rotated 3 times.
Example 2:

Input: nums = [4,5,6,7,0,1,2]
Output: 0
Explanation: The original array was [0,1,2,4,5,6,7] and it was rotated 4 times.
Example 3:

Input: nums = [11,13,15,17]
Output: 11
Explanation: The original array was [11,13,15,17] and it was rotated 4 times. 
 

Constraints:

n == nums.length
1 <= n <= 5000
-5000 <= nums[i] <= 5000
All the integers of nums are unique.
nums is sorted and rotated between 1 and n times.

```python
def findmin(nums):
    nums.sort()
    return nums[0]

print(findmin([3,4,5,1,2]))
print(findmin([4,5,6,7,0,1,2]))
print(findmin([11,13,15,17]))
```

```python
def findmin(nums):
    l = 0
    r = len(nums)-1

    while l < r:
        mid = (l+r)//2
        if nums[mid] < nums[r]:
            r = mid
        else:
            l = mid+1
    return nums[l]

print(findmin([3,4,5,1,2]))
print(findmin([4,5,6,7,0,1,2]))
print(findmin([11,13,15,17]))
```

## Approach 3: Early Exit Optimization
Check if array is already sorted (not rotated) for early exit

```python
def findmin_optimized(nums):
    l = 0
    r = len(nums) - 1
    
    # Early exit: if array is not rotated
    if nums[l] < nums[r]:
        return nums[l]
    
    while l < r:
        mid = (l + r) // 2
        
        if nums[mid] < nums[r]:
            r = mid
        else:
            l = mid + 1
    
    return nums[l]

print(findmin_optimized([3,4,5,1,2]))
print(findmin_optimized([4,5,6,7,0,1,2]))
print(findmin_optimized([11,13,15,17]))  # Early exit case
```

## Approach 4: Compare with Left Boundary
Alternative logic using nums[left] comparison

```python
def findmin_left_compare(nums):
    l = 0
    r = len(nums) - 1
    
    while l < r:
        mid = (l + r) // 2
        
        # If mid element is greater than rightmost, min is in right half
        if nums[mid] > nums[r]:
            l = mid + 1
        # If mid is less than or equal to right, min is in left half (including mid)
        else:
            r = mid
    
    return nums[l]

print(findmin_left_compare([3,4,5,1,2]))
print(findmin_left_compare([4,5,6,7,0,1,2]))
print(findmin_left_compare([11,13,15,17]))
```

## Approach 5: Find Rotation Point
Find where array breaks (nums[i] > nums[i+1])

```python
def findmin_rotation_point(nums):
    n = len(nums)
    if n == 1:
        return nums[0]
    
    l = 0
    r = n - 1
    
    # If not rotated
    if nums[l] < nums[r]:
        return nums[l]
    
    while l <= r:
        mid = (l + r) // 2
        
        # Check if mid is the rotation point
        if nums[mid] > nums[mid + 1]:
            return nums[mid + 1]
        
        # Check if mid-1 is the rotation point
        if nums[mid - 1] > nums[mid]:
            return nums[mid]
        
        # Decide which half to search
        if nums[mid] > nums[0]:
            l = mid + 1
        else:
            r = mid - 1
    
    return nums[0]

print(findmin_rotation_point([3,4,5,1,2]))
print(findmin_rotation_point([4,5,6,7,0,1,2]))
print(findmin_rotation_point([11,13,15,17]))
```

## Complexity Comparison

| Approach | Time | Space | Notes |
|----------|------|-------|-------|
| Sorting | O(n log n) | O(1) | Does not meet requirement |
| Binary Search (compare right) | O(log n) | O(1) | Optimal, cleanest |
| Binary Search + Early Exit | O(log n) | O(1) | Best case O(1) |
| Binary Search (compare left) | O(log n) | O(1) | Alternative logic |
| Find Rotation Point | O(log n) | O(1) | More explicit checks |

```python
def binary_search(nums):
    left = 0
    right = len(nums)-1

    while left < right:

        mid = left + (right - left)//2
        if nums[right] > nums[mid]:
            right = mid
        else:
            left = mid + 1

```
