Given a string s, return the number of palindromic substrings in it.

A string is a palindrome when it reads the same backward as forward.

A substring is a contiguous sequence of characters within the string.

 

Example 1:

Input: s = "abc"
Output: 3
Explanation: Three palindromic strings: "a", "b", "c".
Example 2:

Input: s = "aaa"
Output: 6
Explanation: Six palindromic strings: "a", "a", "a", "aa", "aa", "aaa".

```python
def count_palindrome_substrings(s):
    """Count the number of palindromic substrings in a string."""
    total_count = 0
    
    def expand_around_center(left, right):
        """Expand around center and count palindromes."""
        palindrome_count = 0
        while left >= 0 and right < len(s) and s[left] == s[right]:
            left -= 1
            right += 1
            palindrome_count += 1
        return palindrome_count
    
    for i in range(len(s)):
        # Count odd-length palindromes (center at i)
        odd_palindromes = expand_around_center(i, i)
        # Count even-length palindromes (center between i and i+1)
        even_palindromes = expand_around_center(i, i + 1)
        total_count += odd_palindromes + even_palindromes

    return total_count

print(count_palindrome_substrings("abc"))
print(count_palindrome_substrings("aaa"))
```

```python
def countPalindromesubstring(s):
    count = 0
    def expand(l,r):
        count = 0
        while l >= 0 and r < len(s) and s[l] == s[r]:
            l-=1
            r+=1
            count += 1
        return count
    
    for i in range(len(s)):
        count += expand(i,i)+expand(i,i+1)

    return count

print(countPalindromesubstring("abc"))
print(countPalindromesubstring("aaa"))
```

```python
class Solution:
    def countSubstrings(self, s: str) -> int:
        length = len(s)
        if length <= 1:
            return length
        if len(set(s)) == 1:
            return (1 + length) * length // 2
        count,prev,current = 0,[False] * length,[False] * length
        for i in range(length-1,-1,-1):
            count += 1
            current[i] = True
            for j in range(i+1,length):
                if s[i] == s[j]:
                    if j - i == 1 or prev[j-1]:
                        count += 1
                        current[j] = True
                    else:
                        current[j] = False
                else:
                    current[j] = False
            prev = current[:]
        return count
```

```python
class Solution:
    def countSubstrings(self, s: str) -> int:
        t= "#" + "#".join(s) + "#"
        n = len(t)
    
        p = [0] * n
        center = right = 0
        result = 0
    
        for i in range(n):
            mirror = 2 * center - i
        
            if i < right:
                p[i] = min(right - i, p[mirror])
        
        # Expand around center i
            a = i + p[i] + 1
            b = i - p[i] - 1
        
            while a < n and b >= 0 and t[a] == t[b]:
                p[i] += 1
                a += 1
                b -= 1
        
        # Update center and right boundary
            if i + p[i] > right:
                center = i
                right = i + p[i]
        
            result += (p[i] + 1) // 2
    
        return result
```
