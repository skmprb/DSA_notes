Given an integer array nums, return all the triplets [nums[i], nums[j], nums[k]] such that i != j, i != k, and j != k, and nums[i] + nums[j] + nums[k] == 0.

Notice that the solution set must not contain duplicate triplets.

 

Example 1:

Input: nums = [-1,0,1,2,-1,-4]
Output: [[-1,-1,2],[-1,0,1]]
Explanation: 
nums[0] + nums[1] + nums[2] = (-1) + 0 + 1 = 0.
nums[1] + nums[2] + nums[4] = 0 + 1 + (-1) = 0.
nums[0] + nums[3] + nums[4] = (-1) + 2 + (-1) = 0.
The distinct triplets are [-1,0,1] and [-1,-1,2].
Notice that the order of the output and the order of the triplets does not matter.
Example 2:

Input: nums = [0,1,1]
Output: []
Explanation: The only possible triplet does not sum up to 0.
Example 3:

Input: nums = [0,0,0]
Output: [[0,0,0]]
Explanation: The only possible triplet sums up to 0.

```python
def hastripletsum(nums, target):
    n = len(nums)
    for i in range(n-2):
        st = set()
        for j in range(i, n-1):
            second = target - nums[i] - nums[j]

            if second in st:
                return[nums[i], nums[j], second]
            else:
                st.add(nums[j])
    return []

arr = [-1,0,1,2,-1,-4]
target = 0
print(hastripletsum(arr, target))
```

```python
def gettripletsSum(nums, target):
    # Optimized function to find all unique triplets that sum to target
    n = len(nums)
    nums.sort()  # Sort to enable two-pointer technique and skip duplicates
    result = []
    
    for i in range(n - 2):
        # Skip duplicate values for the first element
        if i > 0 and nums[i] == nums[i - 1]:
            continue
            
        left = i + 1
        right = n - 1
        
        while left < right:
            current_sum = nums[i] + nums[left] + nums[right]
            
            if current_sum == target:
                result.append([nums[i], nums[left], nums[right]])
                
                # Skip duplicates for the second element
                while left < right and nums[left] == nums[left + 1]:
                    left += 1
                # Skip duplicates for the third element
                    while left < right and nums[right] == nums[right - 1]:
                        right -= 1
                    
                left += 1
                right -= 1
            elif current_sum < target:
                left += 1
            else:
                right -= 1
                
    return result

arr = [-1, 0, 1, 2, -1, -4]
target = 0
print(gettripletsSum(arr, target))
```

```python
from typing import List
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        unique = sorted(set(nums))
        freq = {}
        for n in nums:
            freq[n] = freq.get(n, 0) + 1
        
        k = len(unique)
        if k == 0:
            return []
        
        # Early exits
        if unique[0] > 0:      # smallest is positive, no solution
            return []
        if unique[-1] < 0:     # largest is negative, no solution
            return []
        if unique[0] == 0 and unique[-1] == 0:  # only zeros
            return [[0, 0, 0]] if freq[0] >= 3 else []
        
        min_val, max_val = unique[0], unique[-1]
        result = []
        
        for i in range(k):
            a = unique[i]
            if a > 0:
                break
            
            b_min = max(a, -a - max_val)
            b_max = min(-a // 2, -a - min_val)
            
            if b_min > b_max:
                continue
            
            j_start = max(i, bisect_left(unique, b_min))
            j_end = bisect_right(unique, b_max)
            
            for j in range(j_start, j_end):
                b = unique[j]
                
                if a == b and freq[a] < 2:
                    continue
                
                c = -(a + b)
                if c in freq:
                    if c == b and freq[b] < (3 if a == b else 2):
                        continue
                    result.append([a, b, c])
        
        return result
```

```python
from collections import Counter
from typing import List
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        # (x, y, -(x+y))
        nums.sort()
        counter = Counter(nums)
        res = []

        if counter[0] > 2:
            res.append([0] * 3)

        # (-x, -x, -2*x)
        for key in counter.keys():
            if counter[key] > 1 and counter[-2*key] > 0 and key != 0:
                res.append([key, key, -2*key])

        keys_list = list(counter.keys())
        keys_list.sort()
        keys_set = set(keys_list)
        # print(keys_list)
        for i in range(len(keys_list) - 1):
            for j in range(i, len(keys_list)):
                candidate = -(keys_list[i] + keys_list[j])
                if keys_list[i] == keys_list[j]:
                    continue
                
                if candidate <= keys_list[j]:
                    break
                if candidate in keys_set:
                    # print(i, j)
                    res.append([keys_list[i], keys_list[j], candidate])
        return res
```
