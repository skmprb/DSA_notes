Given two strings s1 and s2, return true if s2 contains a permutation of s1, or false otherwise.

In other words, return true if one of s1's permutations is the substring of s2.

 

Example 1:

Input: s1 = "ab", s2 = "eidbaooo"
Output: true
Explanation: s2 contains one permutation of s1 ("ba").
Example 2:

Input: s1 = "ab", s2 = "eidboaoo"
Output: false
 

Constraints:

1 <= s1.length, s2.length <= 104
s1 and s2 consist of lowercase English letters.

```python
def freq(st):
    hash = {}
    for i in st:
        if i in hash:
            hash[i] = hash.get(i)+1
        else:
            hash[i] = 1
    
    return hash

def checkpermutation(st1,st2):
    slide = len(st1)
    start = 0
    end = slide
    freq1 =freq(st1)

    for _ in range(len(st2)-slide-1):
        if freq1 == freq(st2[start:end]):
            return True
        start += 1
        end += 1
    return False


print(checkpermutation("abd", "eidbaooo"))
print(checkpermutation("ab", "eidboaoo"))
```

```python
# This function checks if any permutation of s1 is a substring of s2
def checkpermutation2(s1,s2):
    # Get the lengths of both input strings
    n1 = len(s1)
    n2 =len(s2)

    # If s1 is longer than s2, no permutation of s1 can be a substring of s2
    if n1>n2:
        return False
    
    # Initialize frequency arrays for counting characters (assuming lowercase letters a-z)
    s1_counts = [0] * 26
    s2_counts = [0] * 26

    # Count character frequencies in the first window of s2 (same size as s1) and in s1
    for i in range(n1):
        s1_counts[ord(s1[i]) - 97] += 1  # Convert character to index (a=0, b=1, etc.)
        s2_counts[ord(s2[i]) - 97] += 1
    
    # Check if the first window already matches s1's character frequencies
    if s1_counts == s2_counts:
        return True
    
    # Use sliding window technique to check remaining positions in s2
    for i in range(n1,n2):
        s2_counts[ord(s2[i]) - 97] += 1      # Add new character entering the window
        s2_counts[ord(s2[i-n1]) - 97] -= 1  # Remove character leaving the window
        # Check if current window's character frequencies match s1's frequencies
        if s1_counts == s2_counts:
            return True
    
    # No permutation of s1 found as substring in s2
    return False

print(checkpermutation2("ei", "eidbaooo"))
print(checkpermutation2("ab", "eidboaoo"))
```
