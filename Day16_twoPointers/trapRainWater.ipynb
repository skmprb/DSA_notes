Given n non-negative integers representing an elevation map where the width of each bar is 1, compute how much water it can trap after raining.

 

Example 1:


Input: height = [0,1,0,2,1,0,1,3,2,1,2,1]
Output: 6
Explanation: The above elevation map (black section) is represented by array [0,1,0,2,1,0,1,3,2,1,2,1]. In this case, 6 units of rain water (blue section) are being trapped.
Example 2:

Input: height = [4,2,0,3,2,5]
Output: 9
 

Constraints:

n == height.length
1 <= n <= 2 * 104
0 <= height[i] <= 105

```python
# This function calculates the amount of rainwater that can be trapped between elevation heights
def trapRainWater(list):
    # Initialize left pointer to start from the second element (index 1)
    left = 1
    # Initialize right pointer to start from the second-to-last element
    right = len(list)-2

    # Set initial maximum height on the right side (rightmost element)
    rmax = list[right + 1]
    # Set initial maximum height on the left side (leftmost element)
    lmax = list[left - 1]

    # Variable to accumulate the total water trapped
    watertrapped = 0
    # Continue until the two pointers meet
    while left <= right:
        # If right max is smaller or equal, process from the right side
        if rmax <= lmax:
            # Add water that can be trapped at current right position
            # Water level is limited by rmax, subtract current height
            watertrapped += max(0, rmax-list[right])
            # Update right maximum with current height if it's higher
            rmax = max(rmax, list[right])
            # Move right pointer leftward
            right -= 1
        else:
            # Process from the left side when left max is smaller
            # Add water that can be trapped at current left position
            # Water level is limited by lmax, subtract current height
            watertrapped += max(0, lmax - list[left])
            # Update left maximum with current height if it's higher
            lmax = max(lmax, list[left])
            # Move left pointer rightward
            left += 1
    
    # Return the total amount of water trapped
    return watertrapped

# Test case 1: Expected output is 6 units of water
print(trapRainWater([0,1,0,2,1,0,1,3,2,1,2,1]))
# Test case 2: Expected output is 9 units of water
print(trapRainWater([4,2,0,3,2,5]))
```

```python
from typing import List

class Solution:
    def trap(self, height: List[int]) -> int:
        # Get the length of the height array
        n = len(height)
        # Handle edge case: empty array cannot trap water
        if n == 0:
            return 0
        
        # Initialize two pointers at the start and end of the array
        left, right = 0, n - 1
        # Track the maximum height seen so far from left and right sides
        leftMax, rightMax = 0, 0
        # Accumulate the total trapped water
        trappedWater = 0
        
        # Continue until the two pointers meet
        while left < right:
            # Move the pointer with the smaller height
            if height[left] < height[right]:
                # If left is smaller, check against leftMax
                if height[left] > leftMax:
                    # Update leftMax if current height is higher
                    leftMax = height[left]
                else:
                    # Calculate trapped water: difference between max height and current height
                    trappedWater += leftMax - height[left]
                # Move left pointer inward
                left += 1
            else:
                # If right is smaller, check against rightMax
                if height[right] > rightMax:
                    # Update rightMax if current height is higher
                    rightMax = height[right]
                else:
                    # Calculate trapped water: difference between max height and current height
                    trappedWater += rightMax - height[right]
                # Move right pointer inward
                right -= 1
        
        # Return the total amount of trapped water
        return trappedWater
```

```python
def maxWater(arr):
    # Initialize result variable to store total trapped water
    res = 0

    # For every element of the array (excluding first and last elements)
    # We skip first and last because water cannot be trapped at the boundaries
    for i in range(1, len(arr) - 1):

        # Find the maximum element on its left
        # Start with current element's height
        left = arr[i]
        # Check all elements to the left of current position
        for j in range(i):
            left = max(left, arr[j])

        # Find the maximum element on its right
        # Start with current element's height
        right = arr[i]
        # Check all elements to the right of current position
        for j in range(i + 1, len(arr)):
            right = max(right, arr[j])

        # Update the maximum water
        # Water level at position i is determined by the minimum of left and right max heights
        # Subtract current element's height to get trapped water at this position
        res += (min(left, right) - arr[i])

    return res


if __name__ == "__main__":
    # Test array representing heights of bars
    arr = [2, 1, 5, 3, 1, 0, 4]
    # Print the total amount of trapped rainwater
    print(maxWater(arr))
```

```python
# This function calculates the maximum amount of rainwater that can be trapped
# Logic: For each position, water level = min(max_height_to_left, max_height_to_right) - current_height
def maxWater(arr):
    n = len(arr)

    # left array - stores maximum height seen so far from left to right
    left = [0] * n
    # right array - stores maximum height seen so far from right to left
    right = [0] * n

    # result variable to store total trapped water
    res = 0

    # fill left array - each position stores max height from start to current position
    left[0] = arr[0]
    for i in range(1, n):
        left[i] = max(left[i - 1], arr[i])

    # fill right array - each position stores max height from current position to end
    right[n - 1] = arr[n - 1]
    for i in range(n - 2, -1, -1):
        right[i] = max(right[i + 1], arr[i])

    # calculate the accumulated water element by element
    # water at position i = min(left_max, right_max) - height[i]
    for i in range(1, n - 1):
        min_of_2 = min(left[i], right[i])  # minimum of left and right max heights
        res += min_of_2 - arr[i]  # add trapped water at current position

    return res

if __name__ == "__main__":
    arr = [2, 1, 5, 3, 1, 0, 4]
    print(maxWater(arr))
```
